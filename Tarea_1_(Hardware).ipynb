```python
def leer_archivo_clientes(ruta_archivo):
    clientes = []
    try:
        with open(ruta_archivo, 'r') as archivo:
            # Leer la primera línea para obtener los números de sillas y barberos
            primera_linea = next(archivo)
            partes = primera_linea.strip().split()
            if len(partes) != 3:
                print("Error: La primera línea del archivo debe contener tres números separados por espacios.")
                return []
            num_sillas_espera, num_barberos, num_sillas_barbero = map(int, partes)
            # Leer las líneas restantes para obtener los clientes
            for linea in archivo:
                partes = linea.strip().split()
                if len(partes) == 3:
                    cliente = tuple(map(int, partes))
                    clientes.append(cliente)
                #else:
                    #print(f"Error: La línea '{linea.strip()}' no tiene el formato correcto.")
    except FileNotFoundError:
       print(f"Error: El archivo '{ruta_archivo}' no fue encontrado.")
    return num_sillas_espera, num_barberos, num_sillas_barbero, clientes
```

```python
import time
from threading import Thread, Semaphore

class Hebra(Thread):
    def __init__(self, parent=None, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self.parent = parent

class Barberia:
    def __init__(self, num_sillas_espera, num_barberos, num_sillas_barbero):
        self.num_sillas_espera = num_sillas_espera
        self.num_barberos = num_barberos
        self.num_sillas_barbero = num_sillas_barbero
        self.sillas_espera = Semaphore(num_sillas_espera)
        self.sillas_barbero = Semaphore(num_sillas_barbero)
        self.barbero_disponible = Semaphore(num_barberos)
        self.clientes_esperando = 0

    def llega_cliente(self, num_cliente, espera, corte):
        print(f"Cliente {num_cliente} llega a la barbería.")
        if self.sillas_espera.acquire(blocking=False):
            print(f"Cliente {num_cliente} usa la silla de espera {num_cliente % self.num_sillas_espera + 1}.")
            self.clientes_esperando += 1
            self.sillas_espera.release()
            self.atender_cliente(num_cliente, espera, corte)
        else:
            print(f"Cliente {num_cliente} se va porque no hay sillas disponibles.")

    def atender_cliente(self, num_cliente, espera, corte):
        print(f"Cliente {num_cliente} usa la silla de barbero {num_cliente % self.num_barberos + 1}.")
        with self.barbero_disponible:
            print(f"Barbero {num_cliente % self.num_barberos + 1} atiende a cliente {num_cliente}.")
            self.sillas_barbero.acquire()
            time.sleep(corte)
            self.sillas_barbero.release()
            print(f"Cliente {num_cliente} sale de la barbería (fue atendido).")
            self.clientes_esperando -= 1

def simular_barberia(clientes):
    barberia = Barberia(num_sillas_espera, num_barberos, num_sillas_barbero)
    print("Datos de la barbería:")
    print("Número de sillas de espera:", barberia.num_sillas_espera)
    print("Número de barberos:", barberia.num_barberos)
    print("Número de sillas de barbero:", barberia.num_sillas_barbero)
    for i, cliente in enumerate(clientes):
        time.sleep(cliente[0])
        thread = Hebra(target=barberia.llega_cliente, args=(i, cliente[1], cliente[2]))
        thread.start()

ruta_archivo = "file5.data"  # Ruta al archivo de clientes
num_sillas_espera, num_barberos, num_sillas_barbero, clientes = leer_archivo_clientes(ruta_archivo)
print(clientes)
simular_barberia(clientes)

```
